```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Plotting graphs
file = 'data/Iris_Data.csv'
data = pd.read_csv(file)

plt.plot(data.sepal_width,data.sepal_length, ls='', marker='o')
```

```python
#Multiple Labels under one graph
plt.plot(data.petal_width,data.petal_length,ls='',marker='o', label='Petal')
plt.plot(data.sepal_width,data.sepal_length,ls='',marker='o', label='Sepal')
```

```python
#Histogram
plt.hist(data.sepal_width, bins=25)
```
